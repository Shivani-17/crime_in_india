```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
%matplotlib inline
```

```python
os.listdir('crime-in-india')
```

```python
crime_cases_women = pd.read_csv('crime-in-india/42_Cases_under_crime_against_women.csv')
rape_victims = pd.read_csv('crime-in-india/20_Victims_of_rape.csv')
```

```python
print(crime_cases_women.head())
```

```python
print(crime_cases_women['Group_Name'].value_counts())
```

```python
crime_cases_women[crime_cases_women['Group_Name'] == 'Rape'].groupby('Area_Name')['Group_Name'].count()
```

```python
print(crime_cases_women.columns.values)
```

```python
df_cases_not_investigated = crime_cases_women[crime_cases_women['Group_Name'] == 'Rape'].groupby('Area_Name')['Cases_not_Investigated_or_in_which_investigation_was_refused'].mean().sort_values(ascending=False)
#print(df_cases_not_investigated)
df_cases_not_investigated = df_cases_not_investigated[df_cases_not_investigated > 0]
df_cases_not_investigated.plot(kind='bar',rot=90)
```

```python
print(rape_victims.head())
```

```python
df_rape_victims = rape_victims.groupby('Area_Name')['Rape_Cases_Reported'].mean().sort_values(ascending=False)
#print(df_rape_victims)
df_rape_victims = df_rape_victims[df_rape_victims > 10]
df_rape_victims.plot(kind='bar',rot=90)
```

```python
df_rape_victims_yearly = rape_victims.groupby('Year')['Rape_Cases_Reported'].mean().sort_values(ascending=False)
df_rape_victims_yearly.plot(kind='bar',rot=90)
```

```python
print(rape_victims.columns.values)
```

```python
y = ['Victims_Above_50_Yrs','Victims_Between_10-14_Yrs' ,'Victims_Between_14-18_Yrs'
, 'Victims_Between_18-30_Yrs', 'Victims_Between_30-50_Yrs'
, 'Victims_of_Rape_Total', 'Victims_Upto_10_Yrs']
for item in y :
    df_rape_victims_yearly = rape_victims.groupby('Year')[item].mean()
    plt.xlabel('Year')
    plt.ylabel(item)
    plt.xticks(rotation=90)
    plt.figure()
    df_rape_victims_yearly.plot(kind='bar',rot=90)
    
```

```python
print(rape_victims['Year'].value_counts())
```

```python
y = ['Victims_Above_50_Yrs','Victims_Between_10-14_Yrs' ,'Victims_Between_14-18_Yrs'
, 'Victims_Between_18-30_Yrs', 'Victims_Between_30-50_Yrs'
, 'Victims_of_Rape_Total', 'Victims_Upto_10_Yrs']
i = 0
list_of_df = []
for item in y :
    values = rape_victims.groupby(['Year'])[item].mean()
    list_of_df.append(values.to_frame().reset_index())
#for item in list_of_df :
#    print(item)
final_df = list_of_df[0]
for i in range(1,len(list_of_df)) :
    #print i
    final_df = pd.merge(final_df,list_of_df[i],how='inner',on='Year')
for item in y  :
    plt.figure()
    plt.xlabel('Year')
    plt.ylabel(item)
    plt.plot(final_df['Year'],final_df[item])
```

```python
print(rape_victims.head())
```

```python
area_per_year = rape_victims.groupby(['Area_Name','Year'])['Rape_Cases_Reported'].mean().sort_values(ascending=False).reset_index()
```

```python
plt.xticks(rotation=90)
sns.barplot(x = area_per_year['Area_Name'],y= area_per_year['Rape_Cases_Reported'],hue= area_per_year['Year'])
```

### UP,Bihar and Sikkim observed decrease in the mean number of rape cases reported

```python
area_name = area_per_year['Area_Name'].unique()
for item in area_name :
    plt.figure()
    #plt.xlabel('Year')
    #plt.ylabel('For area ' + item)
    ax = sns.barplot(y = area_per_year[area_per_year['Area_Name'] == item ]['Rape_Cases_Reported'],x =
                     area_per_year[area_per_year['Area_Name'] == item ]['Year'])
    ax.set(xlabel = 'Year', ylabel = item)
    plt.show()
```

```python
area_per_year_withdrawn = crime_cases_women.groupby(['Area_Name','Year'])['Cases_Compounded_or_Withdrawn'].mean().sort_values(ascending=False).reset_index()
plt.xticks(rotation=90)
sns.barplot(x = area_per_year_withdrawn['Area_Name'],y= area_per_year_withdrawn['Cases_Compounded_or_Withdrawn'])
```

###### Madhya Pradesh has highest number of withdrawn cases followed by UP, Maharashtra and Rajasthan

```python
area_name = area_per_year['Area_Name'].unique()
for item in area_name :
    plt.figure()
    ax = sns.barplot(y = area_per_year_withdrawn[area_per_year['Area_Name'] == item ]['Cases_Compounded_or_Withdrawn'],x =
                     area_per_year_withdrawn[area_per_year_withdrawn['Area_Name'] == item ]['Year'])
    ax.set(xlabel = 'Year', ylabel = item)
    plt.show()
```

```python
area_per_year_withdrawn = crime_cases_women.groupby(['Area_Name','Year'])['Cases_Convicted'].mean().sort_values(ascending=False).reset_index()
plt.xticks(rotation=90)
sns.barplot(x = area_per_year_withdrawn['Area_Name'],y= area_per_year_withdrawn['Cases_Convicted'])
```

```python
area_name = area_per_year['Area_Name'].unique()
for item in area_name :
    plt.figure()
    ax = sns.barplot(y = area_per_year_withdrawn[area_per_year['Area_Name'] == item ]['Cases_Convicted'],x =
                     area_per_year_withdrawn[area_per_year_withdrawn['Area_Name'] == item ]['Year'])
    ax.set(xlabel = 'Year', ylabel = item)
    plt.show()
```

##### Tamil Nadu, Madhya Pradesh, Karnataka, Rajasthan among top states where cases were withdrawn by the govt.
